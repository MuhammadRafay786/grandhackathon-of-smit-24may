```python
import requests
import pandas as pd

# Step 1: Fetch SpaceX Launches Data
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Step 2: Normalize JSON and Convert to DataFrame
df = pd.json_normalize(data)

# Step 3: Save to CSV
df.to_csv("spacex_launches.csv", index=False)

print("CSV file saved as 'spacex_launches.csv'")
```

```python
# Q4: Count launches by success flag
success_counts = df['success'].value_counts(dropna=False)
print(success_counts)

# (Optional) Plot a bar chart
success_counts.plot.bar(title='Launch Success vs Failure')
```
